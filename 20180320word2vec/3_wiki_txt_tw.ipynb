```python
# -*- coding: utf-8 -*-
import logging
from gensim.corpora import WikiCorpus
```

# Data preprocessing

```python
%%time
# Number of text
texts_num = 0
# Input file name (https://dumps.wikimedia.org/zhwiki/20180301/zhwiki-20180301-pages-articles.xml.bz2)
wiki_file = "./data/zhwiki-20180301-pages-articles.xml.bz2"

# Setting logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

# Clear content text (wiki xml file)
wiki_corpus = WikiCorpus(wiki_file, dictionary={})

# Output
with open("wiki_texts.txt",'w',encoding='utf-8') as output:
    for text in wiki_corpus.get_texts():
        output.write(' '.join(text) + '\n')
        texts_num += 1
        if texts_num % 10000 == 0:
            logging.info("已處理 %d 篇文章" % texts_num)
```

# Segmentation

```python
# -*- coding: utf-8 -*-
import jieba
```

```python
%%time
# Change to traditional Chinese corpus
jieba.set_dictionary('../word2vec-tutorial/jieba_dict/dict.txt.big')
# Load userdict.txt (https://github.com/rudaoshi/ICTCLASTokenizer/blob/master/src/ictclas/part-of-speech.txt)
"""

"""
jieba.load_userdict("userdict.txt")

# load stopwords set
stopword_set = set()
with open('../word2vec-tutorial/jieba_dict/stopwords.txt','r', encoding='utf-8') as stopwords:
    for stopword in stopwords:
        stopword_set.add(stopword.strip('\n'))
#         
output = open('wiki_seg.txt', 'w', encoding='utf-8')
with open('wiki_zh_tw.txt', 'r', encoding='utf-8') as content:
    for texts_num, line in enumerate(content):
        line = line.strip('\n')
        words = jieba.cut(line, cut_all=False)
        for word in words:
            if word not in stopword_set:
                output.write(word + ' ')
        output.write('\n')
        if (texts_num + 1) % 10000 == 0:
            logging.info("已完成前 %d 行的斷詞" % (texts_num + 1))
output.close()
```

# Training


```python
%%time
from gensim.models import word2vec

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
sentences = word2vec.LineSentence("wiki_seg.txt")
model = word2vec.Word2Vec(sentences, size=250)

#保存模型，供日後使用
model.save("word2vec.model")
```

# Demo

```python
from gensim import models

model = models.Word2Vec.load('./model/word2vec.model')
"""
1. 輸入一個詞，則去尋找前一百個該詞的相似詞
2. 輸入兩個詞，則去計算兩個詞的餘弦相似度
3. 輸入三個詞，進行類比推理 a&b&(!c)-> ?
"""
while True:
    try:
        query = input()
        q_list = query.split()
        
        if q_list[0] == '-1':
            break
        elif len(q_list) == 1:
            print("相似詞前 10 排序")
            res = model.most_similar(q_list[0],topn = 10)
            for item in res:
                print(item[0]+","+str(item[1]))
        elif len(q_list) == 2:
            print("計算 Cosine 相似度")
            res = model.similarity(q_list[0],q_list[1])
            print(res)
        elif len(q_list) == 3:
            print("%s 和 %s，非%s -> " % (q_list[0],q_list[2],q_list[1]))
            res = model.most_similar(positive=[q_list[0],q_list[1]], negative=[q_list[2]], topn= 10)
            for item in res:
                print(item[0]+","+str(item[1]))
        else:
             print("Unknow!!!!")
        print("----------------------------")
        
    except Exception as e:
        print(repr(e))
```
